```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_excel('lagos_state_debt_statistics.xlsx')
except FileNotFoundError:
    print("File not found. Please ensure the file 'lagos_state_debt_statistics.xlsx' is in the same directory.")
    exit()

# Ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows where 'Date' could not be converted
df = df.dropna(subset=['Date'])

# Print column headers
print("DataFrame Columns:")
print(df.columns)

# Set the style
sns.set(style="whitegrid")

# Function to plot line plots
def plot_line(data, x, y, title, xlabel, ylabel):
    print(f"\nPlotting: {title}")
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=data, x=x, y=y, marker='o')
    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()

# Function to plot bar plots
def plot_bar(data, x, y, title, xlabel, ylabel):
    print(f"\nPlotting: {title}")
    plt.figure(figsize=(10, 6))
    sns.barplot(x=x, y=y, data=data, palette='viridis')
    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()

# Function to plot scatter plots
def plot_scatter(data, x, y, title, xlabel, ylabel):
    print(f"\nPlotting: {title}")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, data=data)
    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.tight_layout()
    plt.show()

# Function to calculate correlation
def calculate_correlation(data, col1, col2):
    correlation = data[[col1, col2]].corr()
    print(f"\nCorrelation between {col1} and {col2}:")
    print(correlation)

# 1. Total debt stock over the years
plot_line(df, 'Date', 'Total Debt Stock (NGN)', 'Total Debt Stock of Lagos State Over Time', 'Date', 'Total Debt Stock (NGN)')

# 2. Trend of domestic and external debt over the years
print("\nPlotting: Trend of Domestic and External Debt")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Domestic Debt (NGN)', marker='o', label='Domestic Debt')
sns.lineplot(data=df, x='Date', y='External Debt (NGN)', marker='o', label='External Debt')
plt.title('Trend of Domestic and External Debt', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Debt (NGN)', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 3. Proportion of domestic and external debt
print("\nPlotting: Proportion of Domestic and External Debt")
debt_proportions = df[['Domestic Debt (NGN)', 'External Debt (NGN)', 'Total Debt Stock (NGN)']].iloc[-1]
debt_proportions = debt_proportions / debt_proportions['Total Debt Stock (NGN)']
plt.figure(figsize=(8, 8))
debt_proportions[:-1].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'], fontsize=12)
plt.title('Proportion of Domestic and External Debt', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()

# 4. Debt-to-GDP ratio over the years
plot_line(df, 'Date', 'Debt-to-GDP Ratio', 'Debt-to-GDP Ratio Over Time', 'Date', 'Debt-to-GDP Ratio')

# 5. Average interest rate and inflation rate
interest_rate_avg = df['Interest Rate (%)'].mean()
inflation_rate_avg = df['Inflation Rate (%)'].mean()
print(f'\nAverage Interest Rate: {interest_rate_avg:.2f}%')
print(f'Average Inflation Rate: {inflation_rate_avg:.2f}%')

# 6. Interest rate and inflation rate over the years
print("\nPlotting: Interest Rate vs Inflation Rate")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Interest Rate (%)', marker='o', label='Interest Rate')
sns.lineplot(data=df, x='Date', y='Inflation Rate (%)', marker='o', label='Inflation Rate')
plt.title('Interest Rate vs Inflation Rate', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rate (%)', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 7. Correlation between debt stock and GDP
calculate_correlation(df, 'Total Debt Stock (NGN)', 'GDP (NGN)')

# 8. External debt and exchange rate
print("\nPlotting: External Debt and Exchange Rate")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='External Debt (USD)', marker='o', label='External Debt (USD)')
sns.lineplot(data=df, x='Date', y='Exchange Rate', marker='o', label='Exchange Rate')
sns.lineplot(data=df, x='Date', y='External Debt (NGN)', marker='o', label='External Debt (NGN)')
plt.title('External Debt and Exchange Rate', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 9. Trend of debt stock, GDP, and debt-to-GDP ratio
print("\nPlotting: Debt Stock, GDP, and Debt-to-GDP Ratio")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Total Debt Stock (NGN)', marker='o', label='Total Debt Stock')
sns.lineplot(data=df, x='Date', y='GDP (NGN)', marker='o', label='GDP')
sns.lineplot(data=df, x='Date', y='Debt-to-GDP Ratio', marker='o', label='Debt-to-GDP Ratio')
plt.title('Debt Stock, GDP, and Debt-to-GDP Ratio', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 10. Debt stock, interest rate, and inflation rate
print("\nPlotting: Debt Stock, Interest Rate, and Inflation Rate")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Total Debt Stock (NGN)', marker='o', label='Total Debt Stock')
sns.lineplot(data=df, x='Date', y='Interest Rate (%)', marker='o', label='Interest Rate')
sns.lineplot(data=df, x='Date', y='Inflation Rate (%)', marker='o', label='Inflation Rate')
plt.title('Debt Stock, Interest Rate, and Inflation Rate', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 11. Distribution of total debt stock
print("\nPlotting: Distribution of Total Debt Stock")
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Debt Stock (NGN)'], kde=True, color='skyblue')
plt.title('Distribution of Total Debt Stock', fontsize=14)
plt.xlabel('Total Debt Stock (NGN)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

# 12. Distribution of GDP
print("\nPlotting: Distribution of GDP")
plt.figure(figsize=(10, 6))
sns.histplot(df['GDP (NGN)'], kde=True, color='lightgreen')
plt.title('Distribution of GDP', fontsize=14)
plt.xlabel('GDP (NGN)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

# 13. Distribution of debt-to-GDP ratio
print("\nPlotting: Distribution of Debt-to-GDP Ratio")
plt.figure(figsize=(10, 6))
sns.histplot(df['Debt-to-GDP Ratio'], kde=True, color='salmon')
plt.title('Distribution of Debt-to-GDP Ratio', fontsize=14)
plt.xlabel('Debt-to-GDP Ratio', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

# 14. Total debt stock by year
plot_bar(df, 'Date', 'Total Debt Stock (NGN)', 'Total Debt Stock by Year', 'Date', 'Total Debt Stock (NGN)')

# 15. Relationship between debt stock and interest rate
plot_scatter(df, 'Total Debt Stock (NGN)', 'Interest Rate (%)', 'Debt Stock vs Interest Rate', 'Total Debt Stock (NGN)', 'Interest Rate (%)')

# 16. Relationship between debt stock and GDP per capita
plot_scatter(df, 'GDP per capita (NGN)', 'Total Debt Stock (NGN)', 'GDP per capita vs Debt Stock', 'GDP per capita (NGN)', 'Total Debt Stock (NGN)')

# 17. Relationship between debt stock and population
plot_scatter(df, 'Population', 'Total Debt Stock (NGN)', 'Population vs Debt Stock', 'Population', 'Total Debt Stock (NGN)')

# 18. Trend of debt stock, GDP, and debt-to-GDP ratio, considering population
print("\nPlotting: Debt Stock, GDP, and Debt-to-GDP Ratio, considering Population")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Total Debt Stock (NGN)', marker='o', label='Total Debt Stock')
sns.lineplot(data=df, x='Date', y='GDP (NGN)', marker='o', label='GDP')
sns.lineplot(data=df, x='Date', y='Debt-to-GDP Ratio', marker='o', label='Debt-to-GDP Ratio')
plt.title('Debt Stock, GDP, and Debt-to-GDP Ratio, considering Population', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 19. Average debt-to-GDP ratio
average_debt_to_gdp = df['Debt-to-GDP Ratio'].mean()
print(f'\nAverage Debt-to-GDP Ratio: {average_debt_to_gdp:.2f}')

# 20. Quarterly trend of debt stock
print("\nPlotting: Quarterly Trend of Debt Stock")
df['Quarter'] = df['Date'].dt.to_period('Q')
quarterly_debt = df.groupby('Quarter')['Total Debt Stock (NGN)'].sum().reset_index()
plot_line(quarterly_debt, 'Quarter', 'Total Debt Stock (NGN)', 'Quarterly Trend of Debt Stock', 'Quarter', 'Total Debt Stock (NGN)')

# 21. Monthly trend of debt stock
print("\nPlotting: Monthly Trend of Debt Stock")
df['Month'] = df['Date'].dt.to_period('M')
monthly_debt = df.groupby('Month')['Total Debt Stock (NGN)'].sum().reset_index()
plot_line(monthly_debt, 'Month', 'Total Debt Stock (NGN)', 'Monthly Trend of Debt Stock', 'Month', 'Total Debt Stock (NGN)')

# 22. Correlation between debt stock and inflation rate
calculate_correlation(df, 'Total Debt Stock (NGN)', 'Inflation Rate (%)')

# 23. Debt stock trend by political administration
print("\nPlotting: Debt Stock Trend by Political Administration")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Total Debt Stock (NGN)', marker='o')
plt.title('Debt Stock Trend by Political Administration', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Debt Stock (NGN)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 24. Comparison of debt stock across different states
print("\nPlotting: Comparison of Debt Stock Across Different States")
states_debt = df.groupby('State')['Total Debt Stock (NGN)'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=states_debt, x='State', y='Total Debt Stock (NGN)', palette='coolwarm')
plt.title('Comparison of Debt Stock Across Different States', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Debt Stock (NGN)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 25. Trend of debt servicing costs
plot_line(df, 'Date', 'Debt Servicing Costs (NGN)', 'Trend of Debt Servicing Costs', 'Date', 'Debt Servicing Costs (NGN)')
```
